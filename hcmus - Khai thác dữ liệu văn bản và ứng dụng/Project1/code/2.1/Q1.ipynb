```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_json('Project1_Data.json')

unique_questions = df['question'].nunique()
print("Số lượng câu hỏi: ", unique_questions)

df['question_length'] = df['question'].apply(len)

df['text_length'] = df['text'].apply(len)

df['label'] = df['label'].map({False: 0, True: 1})
label_distribution = df['label'].value_counts()

answer_counts = df['question'].value_counts()
max_answers = answer_counts.max()
question_with_most_answers = answer_counts.idxmax()
print("Số câu trả lời lớn nhất cho một câu hỏi: ", max_answers)
print("Câu hỏi có số câu trả lời lớn nhất: ", question_with_most_answers)

min_answers = answer_counts.min()
question_with_least_answers = answer_counts.idxmin()
print("Số câu trả lời nhỏ nhất cho một câu hỏi: ", min_answers)
print("Câu hỏi có số câu trả lời nhỏ nhất: ", question_with_least_answers)

mean_answers = answer_counts.mean()
print("Trung bình số câu trả lời cho một câu hỏi: ", mean_answers)

plt.figure(figsize=(10, 6))
sns.histplot(df['question_length'], bins=50, kde=True)
plt.title('Phân bố của độ dài câu hỏi')
plt.xlabel('Độ dài câu hỏi')
plt.ylabel('Số lượng')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['text_length'], bins=50, kde=True)
plt.title('Phân bố của độ dài đoạn văn')
plt.xlabel('Độ dài đoạn văn')
plt.ylabel('Số lượng')
plt.show()

plt.figure(figsize=(6, 6))
df['label'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Phân bố của label')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(answer_counts, bins=50)
plt.title('Số lượng câu trả lời cho mỗi câu hỏi')
plt.xlabel('Số lượng câu trả lời')
plt.ylabel('Số lượng câu hỏi')
plt.show()
```
